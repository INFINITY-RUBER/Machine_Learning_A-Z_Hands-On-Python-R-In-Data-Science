# Data Preprocessing Tools

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv('Data.csv')
X = dataset.iloc[:, :-1].values # todas las filas y todas la culumnas menos la ultima
y = dataset.iloc[:, -1].values# todas las filas solo la ultima columna
```

## Taking care of missing data
# Cuidar los datos faltantes

```
from sklearn.impute import SimpleImputer # cargar libreria "SimpleImputer"
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')# asignar el metodo a remplazar los nan por la media (missing_values=np.nan, strategy='mean')
# strategies: ['mean', 'median', 'most_frequent', 'constant']
imputer.fit(X[:, 1:3])# tomamos los vlores numericos (all filas y las columnas 2 y 3)
X[:, 1:3] = imputer.transform(X[:, 1:3])# remplazamos en la mismas pociciones
```

```
print(X)
```

# Encoding categorical data
(Codificación de datos categóricos)

### Encoding the Independent Variable
(Codificación de la variable independiente)

```
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder # libreria de codificacion en caliente
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')# remainder='passthrough'>>> indica que mantenemos las columnas que no se aplico la codificacion
X = np.array(ct.fit_transform(X)) # con numpy agrupamos de nuevo los datos codificados 
```

```
print(X)
```

### Encoding the Dependent Variable
(Codificación de la variable dependiente)

```
from sklearn.preprocessing import LabelEncoder # importamos libreria 
le = LabelEncoder() # cargamos el objeto LabelEncoder()
y0 = y
y = le.fit_transform(y) # aplicamos la transformacion a Y
```

```
print(y0)
print(y)
```

## Splitting the dataset into the Training set and Test set
(División del conjunto de datos en el conjunto de entrenamiento y el conjunto de prueba)

```
from sklearn.model_selection import train_test_split #libreria para dividir los datos
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

* # Standardisation = (x - media(x) ) / (standar deviation(x))

* # Normalisation = (x - min(x) ) / (max(x) - min(x))

```
from sklearn.preprocessing import StandardScaler
sc = StandardScaler() # creamos el objeto de estandarizacion 
X_train[:, 3:] = sc.fit_transform(X_train[:, 3:])# transformamos las columnas 3 y 4 del conjunto de entrenamiento
X_test[:, 3:] = sc.transform(X_test[:, 3:])# transformamos las columnas 3 y 4 del conjunto de testeo
```
